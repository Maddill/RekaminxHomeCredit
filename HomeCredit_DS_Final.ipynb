# **Data Exploration & Viz 1**

Cek trend credit pengguna di data biro

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Mengatur gaya Seaborn
sns.set(style="darkgrid")

# Membagi data menjadi kelompok berdasarkan ID
unique_ids = data_bureau_balance_use['SK_ID_BUREAU'].unique()
id_groups = [unique_ids[i:i+20] for i in range(0, len(unique_ids), 20)]

# Membuat plot untuk setiap kelompok ID
for group in id_groups:
    group_data = data_bureau_balance_use[data_bureau_balance_use['SK_ID_BUREAU'].isin(group)]
    plt.figure(figsize=(12, 8))
    sns.lineplot(x='STATUS', y='MONTHS_BALANCE', data=group_data, hue='SK_ID_BUREAU', palette='tab20', linewidth=2)
    plt.title('History Data for SK_ID_BUREAU')
    plt.xlabel('Status')
    plt.ylabel('Month Balance')
    plt.xticks(rotation=45)
    plt.legend(title='SK_ID_BUREAU', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```
